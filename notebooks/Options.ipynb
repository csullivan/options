```python
import pandas, os
date = '2017-06-07'
spyfiles = sorted([file for file in os.listdir('../spy/') if '.bin' in file and '.h5' not in file and date in file])

data = []
print("Number of files found: "+ str(len(spyfiles)))
for filename in spyfiles:
    chain = pandas.read_pickle('../spy/'+filename)
    chain.set_index('Quote_Time', append=True, inplace=True)
    chain = chain.swaplevel('Strike','Expiry').sort_index(level=0)
    chain = chain.drop('JSON',1)
    data.append(chain)
    
```

```python
combined = pandas.concat(data,axis=0).sort_index(level=0)
```

```python
combined = combined.drop_duplicates().sort_index(level=0)
```

```python
combined.to_pickle('../spy/spy_2017-06-07_full.bin')
```
